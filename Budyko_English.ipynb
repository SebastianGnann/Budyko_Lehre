## **1. Introduction**

<span style="color:DarkRed">**Learning Objective**: *You should be familiar with the key parameters of the climatic water balance and be able to explain the relationship between the water and energy balances.*</span>

### **1.1 Mikhail Ivanovich Budyko (January 20, 1920 - December 10, 2001)**

Between 1955 and 1974, Mikhail Budyko ([Image source](https://en.wikipedia.org/wiki/Mikhail_Budyko#)) studied the components of the land and ocean energy balance as part of his research, which was widely recognized in Russia and beyond and laid the foundation for a new field of meteorology - physical climatology (theory of climate). 

<img src="figures/Mikhail_Ivanovich_Budyko.jpg" alt="drawing" width="200"/>

### **1.2 The Long-Term Water Balance of a Catchment**

The water balance of a catchment can be described as follows:

\begin{equation}
\tag{1}
P + Q_{in} = AET + Q_{out} + \Delta S \label{water_balance}
\end{equation}

where $P$ represents precipitation, $AET$ represents actual evaporation, $Q_{in}$ represents inflow (surface and groundwater), $Q_{out}$ represents outflow or discharge (surface and groundwater), and $\Delta S$ represents the change in storage (all in mm/y).

A catchment is defined as the area that drains toward a point (e.g., a river gauging station), primarily due to topography. Accordingly, a catchment (usually) has no surface inflows. The water that does not evaporate largely leaves the catchment as outflow ($Q$) via a stream.

Budyko further assumes that in large catchments ($\gg$1000 km²), subsurface inflows and outflows via groundwater are negligible, and that over long periods (30 years or more), the change in storage approaches zero. 
The long-term water balance for large catchments is then:

\begin{equation}
\tag{2}
P = AET + Q \label{water_balance_Budyko}
\end{equation}

<img src="figures/water_balance.png" alt="drawing" width=250/>

Equation (2) thus assumes that all water inputs into a catchment originate from precipitation and that all water outputs occur in the form of evaporation or streamflow.

### **1.3 The Energy Balance at the Land Surface**
The energy balance at the land surface can be described (in simplified terms) as follows:

\begin{equation}
\tag{3}
R_n = L + H + G \label{energy_balance}
\end{equation}

where $R_n$ represents net radiation, $L$ represents latent heat, $H$ represents sensible heat, and $G$ represents the ground heat flux (all in W/m²).
Latent heat is the energy used for the evaporation of water, since the transition from liquid water to water vapor requires energy.

<img src="figures/energy_balance.png" alt="drawing" width=100/>

Net radiation thus provides all the energy that is then converted into latent heat, sensible heat, or ground heat flux.

### **1.4 The Relationship Between the Water and Energy Balances**

The latent heat can also be expressed in terms of the actual evaporation $AET$:

\begin{equation}
\tag{4}
L = \lambda_w AET \label{latent_heat}
\end{equation}

where $\lambda_w$ is the enthalpy of evaporation, i.e., the amount of energy required to evaporate a certain amount of water (this depends slightly on the temperature $T$). For example, it takes approximately 2.45 MJ to evaporate one liter of water, roughly the same amount of energy as in a bar of chocolate.

The ground heat flux $G$ is usually very small and can therefore be neglected, resulting in the following equation:

\begin{equation}
\tag{5}
R_n = \lambda_w AET + H \label{energy_and_water_balance}
\end{equation}

Equation (5) thus links the water balance to the energy balance via actual evaporation ($AET$), since both water and energy are required for this process. Equation (5) is thus an important foundation of hydroclimatology.

In addition, we now define potential evaporation $PET$. This is the amount of water that could evaporate if there were always sufficient water available (e.g., over a lake). 
This can be approximated using net radiation (the available energy):

\begin{equation}
\tag {6}
PET = \frac{R_n}{\lambda_w} \label{potential_evaporation}
\end{equation}

There are also other formulas for calculating potential evaporation, but these will not be covered here.

If we now express Equation (5) in terms of potential evaporation, we obtain the following equation:

\begin{equation}
\tag{7}
PET = AET + \frac{H}{\lambda_w} \label{pet_and_water_balance}
\end{equation}

The net radiation, or potential evaporation, that does not contribute to water evaporation is converted into sensible heat.

## **2. The Climatic Water Balance** 

<span style="color:DarkRed">**Learning Objective**: *You should be able to explain what the climatic water balance is and why classifying catchments as water-limited or energy-limited is useful.*</span>

### **2.1 The Aridity Index**
The climatic water balance is defined as $P - PET$.
It is closely related to the so-called aridity index $\phi$:

\begin{equation}
\tag{8}
\phi = \frac{PET}{P} \label{aridity_index}
\end{equation}

### **2.2 Energy- and water-limited regions**
Regions where $P$ is greater than $PET$ (or $\phi < 1$) are referred to as **energy-limited**. 
In energy-limited regions, (almost) all energy can be used ($PET \approx AET$) to evaporate water.

Regions where $P$ is less than $PET$ (or $\phi > 1$) are referred to as **water-limited**.
In such regions, not all energy can be used to evaporate water, since there is not enough water available. 
The energy that is not used to evaporate water is converted into sensible heat $H$.

## **3. The Budyko Curve**

<span style="color:DarkRed">**Learning Objective**: *You should be able to explain the derivation of the Budyko curve and why some catchments fall outside the limits.*</span>

### **3.1 The Energy and Water Limits as the Basis of Budyko’s Theory**

We have seen that the division into water-limited and energy-limited regions allows for initial conclusions about the hydrology of a catchment. Budyko has now investigated these relationships on a large scale across different regions, combining theoretical considerations with observational data.

In energy-limited regions (e.g., humid regions such as Scotland or Sweden), actual evaporation approaches potential evaporation.

\begin{equation}
%\tag{}
AET \rightarrow PET \label{}
\end{equation}

In water-limited regions (e.g., arid regions such as southern Spain or the Sahara), evaporation approaches precipitation.

\begin{equation}
%\tag{}
AET \rightarrow P \label{}
\end{equation}

In addition to the aridity index $\phi$, we now define the so-called evaporation index $\psi$:

\begin{equation}
\tag{9}
\psi = \frac{AET}{P} \label{evaporative_faction}
\end{equation}

For increasingly energy-limited (“humid”) regions, the evaporation index approaches the aridity index, since all available energy is used for evaporation.
For increasingly water-limited (“arid”) regions, the evaporation index approaches 1, since all precipitation evaporates.

### **3.2 Budyko’s original curve* *
If the evaporation index is plotted as a function of the aridity index, the evaporation index for humid regions is limited by the available energy, and for arid regions by the available water. 
 
Budyko plotted the aridity index $\phi$ and the evaporation index $\psi$ of actual catchments against each other and found that the catchments follow a curve.
Based on the results of earlier research by other scientists, he approximated this curve with the following formula:

\begin{equation}
\tag{10}
\psi = [\phi \tanh(1/\phi) (1 - \exp(-\phi))]^{1/2} \label{budyko_equation}
\end{equation}

This formula corresponds to a curve that approaches the energy or water limit and fits well with the catchment regions studied by Budyko, but it is purely empirical. The graph below shows the energy or water limit as well as Budyko’s curve.

<img src="figures/Budyko_curve_en.png" alt="drawing" width="400"/>

### **3.3 Real-world catchment data in the “Budyko region”**
Now let’s examine this using real data. We use 671 catchments from the CAMELS dataset, which was introduced above.
Since there are usually no direct measurements of evaporation in a catchment, we use streamflow data, which we convert to evaporation using Equation (2) ($AET = P - Q$). The evaporation index can then be expressed as a function of $Q$ and $P$.

\begin{equation}
\tag{11}
\psi = \frac{AET}{P} = \frac{P - Q}{P} = 1 - \frac{Q}{P}  \label{evaporative_fraction_using_Q}
\end{equation}


The plot shown below indicates that most data points follow the limits postulated by Budyko. Although the curve does not fit perfectly, it accurately reflects the general trend. These data are independent of the data used by Budyko, suggesting that the Budyko curve is also applicable to other regions and thus has more general validity.

### **4.1 Quantifying the Deviation**
As can be seen, not all catchments lie on the Budyko curve. Aside from uncertainties in the data, what reasons might there be for a deviation from the line? To answer this question, we first quantify the deviations from the curve. 

To do this, we calculate the evaporation index for each catchment based on Budyko’s curve by substituting the catchment’s aridity index into Equation (8).

Now we can compare the actual evaporation index (approximated as $1 - Q/P$) with the calculated evaporation index (Equation 8). We calculate the difference between the two values, known as the Budyko anomaly. The further the value is below the curve, the more negative it is; the further the value is above the curve, the more positive it is. You can see this in the following graph.

```python
# define function that creates a Budyko plot coloured according to Budyko anomalies
def plot_Budyko_anomaly(*args):
    
    #todo: add dict for z name and unit
    
    # create figure
    #fig = plt.figure(figsize=(6, 3), constrained_layout=True)
    #axes = plt.axes()
    
    # create figure
    fig, axes = plt.subplots(figsize=(12, 3), constrained_layout=True, 
                             nrows=1, ncols=2, gridspec_kw={'width_ratios': [2, 2]})
    
    # plot catchments in Budyko space coloured according to Budyko anomalies
    im = axes[0].scatter(df["aridity"], 1-df["runoff_ratio"], s=25, c=df["Budyko_anomaly"], 
                      alpha=0.75, lw=0, cmap='RdBu_r', vmin=-0.3, vmax=0.3) 
    
    # specify axes, labels, etc.
    axes[0].set_xlabel("$PET/P$ [-]")
    axes[0].set_ylabel("$1 - Q/P$ [-]")
    axes[0].set_xlim([0, 5])
    axes[0].set_ylim([-0.2, 1.2])
    helper_fcts.plot_Budyko_limits(df["aridity"], df["runoff_ratio"], axes[0])
    helper_fcts.plot_Budyko_curve(np.linspace(0.01,10,1000), axes[0])   
    #axes.set_xscale('log')
    #cbar = fig.colorbar(im, ax=axes)
    #cbar.set_label("Budyko Abweichung [-]")   
    #cbar.mappable.set_clim(vmin=-0.3,vmax=0.3)
    
    # plot map
    world = gpd.read_file("results/ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp")
    gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.gauge_lon, df.gauge_lat))
    #fig, ax = plt.subplots(figsize=(6, 3))
    world.boundary.plot(ax=axes[1], linewidth=0.5, color='black')
    world.plot(ax=axes[1], color='lightgrey', edgecolor='black', alpha=0.25)
    plot = gdf.plot(ax=axes[1], markersize=10, column=gdf["Budyko_anomaly"], alpha=0.75, cmap='RdBu_r', vmin=-0.3, vmax=0.3)
    norm = Normalize(vmin=-0.3, vmax=0.3)
    sm = plt.cm.ScalarMappable(cmap='RdBu_r', norm=norm)
    sm.set_array([]) 
    cbar = fig.colorbar(sm, ax=axes[1], orientation='vertical', label='Budyko anomaly [-]')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.xlim(-130, -65); plt.ylim(20, 55)
    
# run the function
plot_Budyko_anomaly()

%matplotlib inline
```

# **III. Proposed Solutions and Additional Information**


## 1. Introduction

<span style="color:Indigo">**Simple question**: *Two catchments receive the same amount of precipitation but different amounts of net radiation. Which of the catchments has a higher runoff, and how can this be explained?*</span>

Higher net radiation leads to higher potential evaporation. Especially when sufficient water is available, this results in higher actual evaporation and thus a lower streamflow volume. 

<span style="color:Indigo">**Difficult Question**: *These days, there is frequent discussion of the “sponge city” concept: the aim is to retain more water within cities. What effects does an increase in available water have on the conversion of net radiation (Equation 6), and how does this influence the urban climate (e.g., temperature)?*</span>

If more water is available, more net radiation can be converted into latent heat. Since latent heat, unlike sensible heat, does not raise the temperature (but only increases humidity), the temperature in the city remains lower. This effect is particularly noticeable in the summer in forests, which are often significantly cooler than their surroundings, as evaporation, similar to human sweating, has a cooling effect.

## 2. The Climatic Water Balance

<span style="color:Indigo">**Simple question**: *Where do we find higher streamflow: in water-limited or energy-limited regions? Why is that?*</span>

Water-limited regions have lower streamflow than energy-limited regions, since more precipitation evaporates there and thus does not become streamflow.

<span style="color:Indigo">**Difficult question**: *What is the lowest value that streamflow can be in water- or energy-limited regions? Why is that?*</span>

In water-limited regions, streamflow often approaches 0, since almost all precipitation evaporates here. 
In energy-limited regions, on the other hand, streamflow often approaches the 1:1 line, i.e., $Q = P - PET$. Here, there is sufficient water available so that (almost) all energy is used to evaporate water ($PET \approx AET$). However, actual evaporation can never exceed potential evaporation; therefore, the discharge in energy-limited regions is never less than the climatic water balance.

The relationship is, however, never perfect. For one thing, environmental data is always subject to uncertainty. For another, there are additional factors beyond the climatic water balance that influence streamflow generation (see below).

## 3. The Budyko Curve

<span style="color:Indigo">**Question**: *Some catchments lie outside the theoretical boundaries (in the gray area). What could be the reasons for this? Name at least two.*</span>

Reasons for this may include uncertainties in the data. Example: In mountainous regions, precipitation measurement is difficult, and often less $P$ is measured than actually falls. This can result in the measured streamflow being greater than the measured precipitation.

In some catchments, the assumption that all water is supplied by precipitation does not hold true. Example: groundwater flows in and contributes additional streamflow (or evaporation).

In some catchments, not all water leaves the catchment as streamflow or evaporation. Example: groundwater flows out of the catchment underground.

Additionally, human influences can disrupt the water balance, e.g., water withdrawal. Although the CAMELS US data represent “natural” catchments, human influences cannot be entirely ruled out.

## 5. The Water Balance for Different Years

<span style="color:Indigo">**Simple Question**: *If we look at shorter time periods, such as individual years, certain assumptions underlying the Budyko curve no longer hold.
What are those assumptions? And how does this manifest in the annual data?*</span>

If we look at long-term averages, e.g., over 30 years (“climatological mean”), it becomes more likely that the change in storage in the catchment actually approaches zero. In individual years, this change can be very large and thus lead to deviations from the Budyko curve. This is clearly visible in the example data. The long-term average is often quite close to the Budyko curve, whereas the annual averages are scattered and usually deviate more significantly from the curve. 

<span style="color:Indigo">**Tough question**: *Why does the evaporation index fluctuate very strongly from year to year in some catchments but not in others?*</span>



In regions that are consistently water-limited (e.g., Santa Cruz C near Santa Ynez, CA), nearly all precipitation evaporates each year, and the evaporation index is almost always close to 1, regardless of how high the aridity index is. This clearly indicates that a catchment is water-limited. 

In regions that are constantly energy-limited (e.g., Pine Creek at Cedar Run, PA, or ALSEA RIVER NEAR TIDEWATER, OR), the evaporation index is limited by the aridity index. Evaporation can never exceed the amount of energy available. However, not all available energy is utilized (i.e., $1 - Q/P < PET/P$). One reason for this is the seasonality of precipitation and potential evaporation. Potential evaporation is highest in summer, but summer is often drier than winter—precipitation and evaporation are thus out of phase. As a result, not all precipitation can be evaporated, even if sufficient energy is available on an annual average.

In catchments that alternate from water-limited to energy-limited conditions from year to year (e.g., Garcitas Creek near Inez, TX), it can be observed that in severely water-limited years, the evaporation index is close to 1, whereas in energy-limited years, it is constrained by the aridity index and is less than 1.

However, there are other catchment characteristics that can influence behavior from year to year. WOLF RIVER AT LANGLADE, WI, for example, is a river fed primarily by productive aquifers. These stabilize the flow from year to year, acting like a natural reservoir. As a result, annual fluctuations in precipitation are buffered, and both streamflow and evaporation remain fairly constant.

Some catchments behave quite differently, such as BUFFALO FORK AB LAVA CREEK NEAR MORAN, WY, where approximately 70% of precipitation falls as snow. Snow can significantly influence the water and energy balance (e.g., by reflecting incoming radiation), a topic that is still being researched and is of enormous importance in relation to climate change.
